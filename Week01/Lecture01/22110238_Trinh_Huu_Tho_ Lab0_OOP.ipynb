## Bài 1 - Opperator Overloading

Định nghĩa lớp `Date` để biểu diễn thời gian ngày, tháng, năm trong năm. Lớp này cung cấp một số phương thức sau:

- Định nghĩa hàm `__init__(self, day, month, year)` để khởi tạo một đối tượng thời gian với `ngày (day)`, `tháng (month)`, và `năm (year)` cho trước (ta sẽ gọi hàm bằng cách viết `Date(day, month, year)`).

- Định nghĩa hàm `__str__(self)` để trả về chuỗi `day/month/year` mô tả một `Date object` (ta sẽ gọi hàm bằng cách viết `print(date)`).

- Định nghĩa hàm `__eq__(self, d)` để kiểm tra ngày đang xét có bằng ngày `d`? Ta sẽ gọi hàm bằng cách viết `d1 == d2`)

- Định nghĩa hàm `__ne__(self, d)` để kiểm tra ngày đang xét và `d` có khác nhau? Ta sẽ gọi hàm bằng cách viết `d1 != d2`)?

- Định nghĩa hàm `__gt__(self, d)` để kiểm tra ngày đang xét có lớn hơn `d`? Ta sẽ gọi hàm bằng cách viết `d1 > d2`.

- Định nghĩa hàm `__lt__(self, d)` để kiểm tra ngày đang xét có bé hơn `d`? Ta sẽ gọi hàm bằng cách viết `d1 < d2`.

- Định nghĩa hàm `__ge__(self, d)` để kiểm tra ngày đang xét có lớn hơn hoặc bằng `d`.? Ta sẽ gọi hàm bằng cách viết `d1 >= d2`.

- Định nghĩa hàm `__le__(self, d)` để kiểm tra ngày đang xét có bé hơn hoặc bằng `d`? Ta sẽ gọi hàm bằng cách viết `d1 <= d2`.

- Định nghĩa hàm `dayOfWeek(self)` trả về số nguyên mô tả ngày thứ mấy trong tuần của ngày hiện tại (giả định thứ 2 ứng với 1, thứ 3 ứng với 2, ..., chủ nhật ứng với 7). Ta sẽ gọi hàm bằng cách viết `d.dayOfWeek()`. Ví dụ: giả sử `d = Date(14, 4, 2021)` thì `d.dayOfWeek()` trả về 3, tức là ngày `thứ 4`. Tham khảo công thức tính xác định thứ trong tuần dựa vào ngày/tháng/năm ở https://en.wikipedia.org/wiki/Determination_of_the_day_of_the_week.

- (**\***) Định nghĩa hàm `dayDiff(self, d)` trả về số ngày giữa ngày hiện tại và ngày `d`. Ta sẽ gọi hàm bằng cách viết `d1.dayDiff(d2)`.

- (**\***) Định nghĩa hàm `nextDate(self)` trả về ngày kế sau ngày hiện tại một ngày. Ta sẽ gọi hàm bằng cách viết `d.nextDate()`.


```python
import datetime


class Date:
    """Simple Date class with comparison operators and utilities."""

    def __init__(self, day, month, year):
        self.day = int(day)
        self.month = int(month)
        self.year = int(year)
        # validate date by constructing a datetime.date (will raise ValueError if invalid)
        datetime.date(self.year, self.month, self.day)

    def __str__(self):
        return f"{self.day}/{self.month}/{self.year}"

    def _ymd(self):
        return (self.year, self.month, self.day)

    def __eq__(self, d):
        if not isinstance(d, Date):
            return NotImplemented
        return self._ymd() == d._ymd()

    def __ne__(self, d):
        eq = self.__eq__(d)
        if eq is NotImplemented:
            return NotImplemented
        return not eq

    def __lt__(self, d):
        if not isinstance(d, Date):
            return NotImplemented
        return self._ymd() < d._ymd()

    def __le__(self, d):
        if not isinstance(d, Date):
            return NotImplemented
        return self._ymd() <= d._ymd()

    def __gt__(self, d):
        if not isinstance(d, Date):
            return NotImplemented
        return self._ymd() > d._ymd()

    def __ge__(self, d):
        if not isinstance(d, Date):
            return NotImplemented
        return self._ymd() >= d._ymd()

    def dayOfWeek(self):
        """Return weekday as integer: Monday=1, ..., Sunday=7."""
        # Sakamoto's algorithm: returns 0=Sunday,1=Monday,...,6=Saturday
        y = self.year
        m = self.month
        d = self.day
        t = [0, 3, 2, 5, 0, 3, 5, 1, 4, 6, 2, 4]
        y0 = y
        if m < 3:
            y0 -= 1
        w = (y0 + y0 // 4 - y0 // 100 + y0 // 400 + t[m - 1] + d) % 7
        # map 0 (Sunday) -> 7, others keep as-is (Monday=1,...)
        return 7 if w == 0 else w

    def dayDiff(self, d):
        """Return the absolute number of days between self and d."""
        if not isinstance(d, Date):
            raise TypeError("dayDiff expects a Date")
        dt1 = datetime.date(self.year, self.month, self.day)
        dt2 = datetime.date(d.year, d.month, d.day)
        return abs((dt1 - dt2).days)

    def nextDate(self):
        """Return a new Date that is one day after self."""
        dt = datetime.date(self.year, self.month, self.day) + datetime.timedelta(days=1)
        return Date(dt.day, dt.month, dt.year)
```

## Bài 2 - Inheriance

- Tạo lớp `Automobile` gồm các thuộc tính sau:

  - `__make`: hãng sản xuất

  - `__year`: năm sản xuất

  - `__mileage`: số mile đi được trên một đơn vị nhiên liệu

  - `__price`: giá tiền

- Định nghĩa phương thức `__init__` để khởi tạo các thuộc tính trên.

- Định nghĩa phương thức `__str__` để trả về chuỗi gồm các thuộc tính trên để cung cấp thông tin về một `automobile` cụ thể.

- Tạo các lớp `Car`, `Truck`, `SUV` kế thừa lớp `Automobile`. Trong đó:

  - Lớp `Car` bổ sung thêm thuộc tính `__doors` (số cửa sổ)

  - Lớp `Truck` bổ sung thêm thuộc tính `__drive_type` (loại)

  - Lớp `SUV` bổ sung thêm thuộc tính `__pass_cap` (sức chứa)

- Định nghĩa phương thức `__init__` tương ứng cho các lớp `Car`, `Truck`, `SUV`. Phương thức này nên gọi đến phương thức `__init__` của lớp `Automobile`.

- Định nghĩa các phương thức `__str__` để trả về chuỗi gồm các thuộc tính tương ứng để cung cấp thông tin về một `Car`, `Truck`, và `SUV` cụ thể.

- Tạo một vài `object` của các lớp `Automobile`, `Car`, `Truck`, và `SUV` với các thuộc tính cụ thể.

- Sử dụng các `property` với `decorator` syntax để truy xuất và gán giá trị cho các thuộc tính của các `object` đã tạo. In các thông tin này ra màn hình.

Class Diagram cho các lớp được mô tả ở hình sau:

![Automobile Class Diagram](./image/AutomobileClassDiagram.PNG)


```python
# Automobile base class and subclasses Car, Truck, SUV
class Automobile:
    def __init__(self, make, year, mileage, price):
        # private attributes (name-mangled)
        self.__make = make
        self.__year = int(year)
        self.__mileage = float(mileage)
        self.__price = float(price)

    def __str__(self):
        return f"Automobile(make={self.__make}, year={self.__year}, mileage={self.__mileage}, price={self.__price})"

    # properties for access and assignment
    @property
    def make(self):
        return self.__make

    @make.setter
    def make(self, value):
        self.__make = value

    @property
    def year(self):
        return self.__year

    @year.setter
    def year(self, value):
        self.__year = int(value)

    @property
    def mileage(self):
        return self.__mileage

    @mileage.setter
    def mileage(self, value):
        self.__mileage = float(value)

    @property
    def price(self):
        return self.__price

    @price.setter
    def price(self, value):
        self.__price = float(value)


class Car(Automobile):
    def __init__(self, make, year, mileage, price, doors):
        super().__init__(make, year, mileage, price)
        self.__doors = int(doors)

    @property
    def doors(self):
        return self.__doors

    @doors.setter
    def doors(self, value):
        self.__doors = int(value)

    def __str__(self):
        return f"Car(make={self.make}, year={self.year}, mileage={self.mileage}, price={self.price}, doors={self.__doors})"


class Truck(Automobile):
    def __init__(self, make, year, mileage, price, drive_type):
        super().__init__(make, year, mileage, price)
        self.__drive_type = drive_type

    @property
    def drive_type(self):
        return self.__drive_type

    @drive_type.setter
    def drive_type(self, value):
        self.__drive_type = value

    def __str__(self):
        return f"Truck(make={self.make}, year={self.year}, mileage={self.mileage}, price={self.price}, drive_type={self.__drive_type})"


class SUV(Automobile):
    def __init__(self, make, year, mileage, price, pass_cap):
        super().__init__(make, year, mileage, price)
        self.__pass_cap = int(pass_cap)

    @property
    def pass_cap(self):
        return self.__pass_cap

    @pass_cap.setter
    def pass_cap(self, value):
        self.__pass_cap = int(value)

    def __str__(self):
        return f"SUV(make={self.make}, year={self.year}, mileage={self.mileage}, price={self.price}, pass_cap={self.__pass_cap})"
```

```python
# test your code here
# Create objects from each class
a = Automobile("Toyota", 2010, 30.5, 15000)
c = Car("Honda", 2018, 35.0, 20000, 4)
t = Truck("Ford", 2015, 15.0, 30000, "4x4")
s = SUV("BMW", 2020, 25.0, 50000, 5)

# Print initial string representations (uses __str__)
print("Initial objects:")
print(a)
print(c)
print(t)
print(s)

print("\nProperties (read access):")
print(
    f"Automobile make: {a.make}, year: {a.year}, mileage: {a.mileage}, price: ${a.price:,.2f}"
)
print(f"Car doors: {c.doors}")
print(f"Truck drive type: {t.drive_type}")
print(f"SUV passenger capacity: {s.pass_cap}")

# Modify some properties using setters
a.price = 14000
c.doors = 2
t.drive_type = "RWD"
s.pass_cap = 7

print("\nAfter updates (using setters):")
print(
    f"Automobile: make={a.make}, year={a.year}, mileage={a.mileage}, price=${a.price:,.2f}"
)
print(f"Car: make={c.make}, doors={c.doors}, price=${c.price:,.2f}")
print(f"Truck: make={t.make}, drive_type={t.drive_type}, price=${t.price:,.2f}")
print(f"SUV: make={s.make}, pass_cap={s.pass_cap}, price=${s.price:,.2f}")
```

## Bài 3 - Polymorphism

- Tạo lớp `Animal` gồm thuộc tính `__species` (loài).

- Định nghĩa phương thức `__init__` cho lớp `Animal` để khởi tạo thuộc tính này.

- Định nghĩa phương thức `__str__` để trả về chuỗi gồm thuộc tính trên để cung cấp thông tin về một `animal` cụ thể.

- Định nghĩa phương thức `speak` để trả về chuỗi mô tả tiếng kêu của một `animal` cụ thể.

- Tạo các lớp `Dog`, `Cat`, `Bird`, `Fish` kế thừa lớp `Animal`. Overide lại các phương thức `__init__`, `__str__`, và `speak` một cách phù hợp.

- Tạo một danh sách các `object` của các lớp `Animal`, `Dog`, `Cat`, `Bird` và `Fish` với các giá trị thuộc tính cụ thể.

- Duyệt qua danh sách để in ra thông tin của các động vật và tiếng kêu của chúng.


```python
# write your code here


class Animal:
    """Base Animal class with a private __species attribute."""

    def __init__(self, species):
        self.__species = species

    def __str__(self):
        return f"Animal(species={self.__species})"

    def speak(self):
        return "<unknown sound>"

    @property
    def species(self):
        return self.__species


class Dog(Animal):
    def __init__(self, name):
        super().__init__("Dog")
        self.__name = name

    def __str__(self):
        return f"Dog(name={self.__name}, species={self.species})"

    def speak(self):
        return "Woof!"

    @property
    def name(self):
        return self.__name


class Cat(Animal):
    def __init__(self, name):
        super().__init__("Cat")
        self.__name = name

    def __str__(self):
        return f"Cat(name={self.__name}, species={self.species})"

    def speak(self):
        return "Meow!"

    @property
    def name(self):
        return self.__name


class Bird(Animal):
    def __init__(self, name):
        super().__init__("Bird")
        self.__name = name

    def __str__(self):
        return f"Bird(name={self.__name}, species={self.species})"

    def speak(self):
        return "Tweet!"

    @property
    def name(self):
        return self.__name


class Fish(Animal):
    def __init__(self, name):
        super().__init__("Fish")
        self.__name = name

    def __str__(self):
        return f"Fish(name={self.__name}, species={self.species})"

    def speak(self):
        # Fish are typically quiet — use a small descriptive string
        return "... (blub)"

    @property
    def name(self):
        return self.__name
```

```python
# test your code here

animals = [
    Animal("Unknown"),
    Dog("Buddy"),
    Cat("Whiskers"),
    Bird("Tweety"),
    Fish("Goldie"),
]

for a in animals:
    # polymorphic behavior: __str__ and speak() are resolved on the instance
    print(str(a))
    print("->", a.speak())
    print()
```
