### 2. Tạo thư mục (directory) trong HDFS

-  Xem nội dung của directory trong HDFS

```python
!/usr/local/hadoop-3.3.6/bin/hdfs dfs -ls
```

- Xem root directory trong HDFS

```python
!/usr/local/hadoop-3.3.6/bin/hdfs dfs -ls /
```

- Tạo thư mục tên `test` trong HDFS

```python
# YOUR CODE
!/usr/local/hadoop-3.3.6/bin/hdfs dfs -mkdir test
```

- Kiểm tra lại xem thư mục đã tạo thành công chưa

```python
!/usr/local/hadoop-3.3.6/bin/hdfs dfs -ls
```

- Tạo các thư mục con theo cấu trúc sau:
  - test/test1
  - test/test2/test3

```python
# YOUR CODE
!/usr/local/hadoop-3.3.6/bin/hdfs dfs -mkdir test/test1
!/usr/local/hadoop-3.3.6/bin/hdfs dfs -mkdir -p test/test2/test3
```

- Kiểm tra lại xem thư mục đã tạo thành công chưa

```python
!/usr/local/hadoop-3.3.6/bin/hdfs dfs -ls
```


- Kết quả cho thấy rằng bạn chỉ thấy thư mục `test`. Để xem một cách đệ quy nội dung của một thư mục, hãy sử dụng `-ls -R`

```python
!/usr/local/hadoop-3.3.6/bin/hdfs dfs -ls -R
```

### 3. Xóa thư mục

- Xóa thư mục `test2` một cách đệ quy (xóa cả các thư mục con) dùng lệnh `-rm -R`

```python
# YOUR CODE
!/usr/local/hadoop-3.3.6/bin/hdfs dfs -rm -R test/test2
```

- Kiểm tra lại xem thư mục được xóa thành công hay chưa

```python
!/usr/local/hadoop-3.3.6/bin/hdfs dfs -ls -R
```

### 4. Upload một file vào HDFS

- Tải file

```python
!wget http://files.grouplens.org/datasets/movielens/ml-100k.zip
```

- Extract file

```python
!unzip /content/ml-100k.zip -d /content/ml-100k_folder
```

- Tạo folder trong HDFS tên `ml-100k`

```python
# YOUR CODE
!/usr/local/hadoop-3.3.6/bin/hdfs dfs -mkdir ml-100k
```

- Copy file từ local tại `ml-100k_folder/ml-100k/` vào thư mục của HDFS tên `ml-100k` vừa mới tạo

```python
# YOUR CODE
!/usr/local/hadoop-3.3.6/bin/hdfs dfs -put -f /content/ml-100k_folder/ml-100k/* ml-100k
```

- Liệt kê các file trong mục của HDFS tên `ml-100k`

```python
# YOUR CODE
!/usr/local/hadoop-3.3.6/bin/hdfs dfs -ls ml-100k
```

- Xem vài dòng đầu của file `ml-100k/u.data` với `-head`

```python
# YOUR CODE
!/usr/local/hadoop-3.3.6/bin/hdfs dfs -head ml-100k/u.data
```

- Xem vài dòng cuối của file `ml-100k/u.data` với `-tail`

```python
# YOUR CODE
!/usr/local/hadoop-3.3.6/bin/hdfs dfs -tail ml-100k/u.data
```

- Mỗi dòng gồm một bộ (`userID`, `movieID`, `rating`, `timestamp`) mô tả một `userid` đánh giá (`rating`) một `movieid` vào thời gian (`timestamp`) nào.

### 5. Lấy file từ HDFS ra local

- Dùng `-get` để copy thư file `ml-100k/u.data` từ HDFS ra local ở thư mục `test`

```python
# YOUR CODE
!/usr/local/hadoop-3.3.6/bin/hdfs dfs -get ml-100k/u.data test/
```
