### Câu 1

Cho một hóa đơn sách gồm các thành phần (MaHD, TenSach_TacGia, SoLuong, DonGia) như sau:

```python
orders = [ ["34587", "Learning Python, Mark Lutz", 4, 40.95], 
           ["98762", "Programming Python, Mark Lutz", 5, 56.80], 
           ["77226", "Head First Python, Paul Barry", 3,32.95],
           ["88112", "Einführung in Python3, Bernd Klein", 3, 24.99]]
```

Viết một chương trình Python trả về một danh sách các bộ 2 gồm `MaHD` và `SoTien` là tích của giá mỗi mặt hàng và số lượng. Sản phẩm sẽ được tăng thêm 10 đồng nếu giá trị đơn hàng nhỏ hơn 100 đồng. Viết một chương trình Python sử dụng lambda function và `map`.

```python
# write your code here
mapped_orders = map(lambda x: [x[0], x[3] + 10 if x[3] < 100 else x[3]], orders)
print(list(mapped_orders))
```

### Câu 2

Cho một tập các đơn hàng có dạng [MaHD, (MaSP, SoLuong, DonGia), ... (MaSP, SoLuong, DonGia)] như sau:

```python
orders = [ [1, ("5464", 4, 9.99), ("8274",18,12.99), ("9744", 9, 44.95)], 
           [2, ("5464", 9, 9.99), ("9744", 9, 44.95)],
           [3, ("5464", 9, 9.99), ("88112", 11, 24.99)],
           [4, ("8732", 7, 11.99), ("7733",11,18.99), ("88112", 5, 39.95)] ]
```

Viết một chương trình trả về một danh sách gồm các bộ 2 với (MaHD, TongTien), với TongTien là tổng số tiền của mỗi đơn hàng.

```python
# write your code here
def order_total(order):
    # order is [MaHD, (MaSP, SoLuong, DonGia), ...]
    # return (MaHD, TongTien) where TongTien = sum(SoLuong * DonGia) for all items in the order
    return (order[0], sum(item[1] * item[2] for item in order[1:]))

mapped_orders = list(map(order_total, orders))
print(mapped_orders)
```
